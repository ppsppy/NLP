### 分词器jieba的使用及高频词提取实战

分词时有三种模式：全模式，搜索模式，精确模式

通过自定义字典来删除一切无意义的字词

通过往字典中加入新词来使分词更精确

```python
import jieba
```

```python
sent='贫穷的纠缠，病魔的折磨他毫不在意，只是后悔从来没有光顾过他的意识。作品表现了天才、个性与物质文明以及现代婚姻、家庭生活之间的矛盾，有着广阔的生命视角，用散发着消毒水味道的手术刀对皮囊包裹下的人性进行了犀利地解剖，混合着看客讪笑的幽默和残忍的目光。'
```

默认分词（精确模式）

```python
seg_list=jieba.cut(sent)
print('/'.join(seg_list))
```

精确模式

```python
seg_list=jieba.cut(sent,cut_all=False)
print('/'.join(seg_list))
```

全模式

```python
seg_list=jieba.cut(sent,cut_all=True)
print('/'.join(seg_list))
```

搜索模式

```python
seg_list=jieba.cut_for_search(sent)
print('/'.join(seg_list))
```

创造一个字典，存储一些需要过滤的词

```python
dictionary=['的','，','。','了','、','也','上','“','”','和','对','在','他','是','中']
```

```python
for word in dictionary:
    if c[word]:
        del c[word]
```

```python
c.most_common(10)
```
